```python
import requests
```

```python
# getting title
url_string = "https://scholar.google.com/scholar?start=0&q=(%22Machine+Learning%22%7C%22Deep+Learning%22)(%22Radiation+oncology%22%7C%22cancer+detection%22)&hl=en&as_sdt=0,44"
result = requests.get(url_string)
```

```python
type(result)
```

```python
# calling .text method on the result variable returns the html document
result.content
```

```python
# to extract specific information from the text file we can use bs4
import bs4
```

```python
soup = bs4.BeautifulSoup(result.content, "lxml")
```

```python
print(soup.prettify())
```

```python
# create lists to store titles, year, citations, url, journal, etc.
titles = []
year = []
citations = []
journal = []
authors = []
```

```python
# each result is encapsulated by a [data-lid] tag
soup.select('[data-lid]')[0]    # returns a list
```

```python
# grabbinng title
(soup.select('[data-lid]')[0].select('h3'))[0].get_text()
```

```python
# grabbing url
soup.select('[data-lid]')[0].find_all('a', href = True)[0]['href']
```

```python
# grabbing author names
auth_list = []
for author in soup.select('.gs_a')[0].find_all('a'):
    auth_list.append(author.get_text())
    
', '.join(auth_list)   # A string of author names
```

```python
# grabbing year
for value in soup.select('.gs_a')[2].get_text().split():
    if value.isnumeric():
        print(value)
```

```python
# grabbing journal name as it is typically the last item
soup.select('.gs_a')[2].get_text().split()[-1]
```

```python
# grabbing citations
# the 'gs_fl' class containing citation for each search item = (i*2)+1 ( i = 0, the index is 1)
soup.select('.gs_fl')[1].find_all('a')[2].get_text()
```
